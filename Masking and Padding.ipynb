```python
def create_mask(mask_size, numeric_values):
    mask = [[0] * mask_size for _ in range(mask_size)]

    for i in range(mask_size):
        for j in range(mask_size):
            mask[i][j] = numeric_values[i][j]

    return mask

# Get input from the user
mask_size = int(input("Enter the mask size: "))

numeric_values = []
print("Enter numeric values for the mask:")
for i in range(mask_size):
    row = []
    for j in range(mask_size):
        value = int(input(f"Enter value for position ({i+1},{j+1}): "))
        row.append(value)
    numeric_values.append(row)

# Create the mask
mask = create_mask(mask_size, numeric_values)

# Display the mask
print("\nGenerated Mask:")
for row in mask:
    print(row)
```

```python
import matplotlib.pyplot as plt
def add_padding(image_path, padding_size, padding_type='zeros'):

    # Read the image from Google Drive
    image = cv2.imread(image_path)

    # Check if the image is successfully loaded
    if image is None:
        print("Error: Could not read the image.")
        return

    # Apply padding based on user-specified type
    if padding_type == 'zeros':
        padded_image = cv2.copyMakeBorder(image, padding_size, padding_size, padding_size, padding_size, cv2.BORDER_CONSTANT, value=0)
    elif padding_type == 'copy':
        padded_image = cv2.copyMakeBorder(image, padding_size, padding_size, padding_size, padding_size, cv2.BORDER_REPLICATE)
    else:
        print("Error: Invalid padding type. Supported types are 'zeros' or 'copy'.")
        return

    # Display the original and padded images
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Padded Image")
    plt.imshow(cv2.cvtColor(padded_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.show()

    return padded_image

# Example usage:
image_path_in_drive = "/monkey.jpg"  # Replace with the actual image path in Google Drive
padding_size = 20
padding_type = 'zeros'  # Change to 'copy' if desired

padded_image = add_padding(image_path_in_drive, padding_size, padding_type)
```
