rotating an image

```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

uploaded = files.upload()

if len(uploaded) > 0:

    file_name = next(iter(uploaded))

    image = cv2.imread(file_name)

    angle = 60

    height, width = image.shape[:2]

    rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), angle, 1)

    rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))

    cv2_imshow(image)
    cv2_imshow(rotated_image)

    rotated_file_name = 'rotated_' + file_name
    cv2.imwrite(rotated_file_name, rotated_image)
else:
    print("No image uploaded.")
```

Resizing an image

```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

uploaded = files.upload()

if len(uploaded) > 0:
    file_name = next(iter(uploaded))

    image = cv2.imread(file_name)

    original_height, original_width = image.shape[:2]

    new_width = int(input("Enter the new width: "))
    new_height = int(input("Enter the new height: "))

    resized_image = cv2.resize(image, (new_width, new_height))

    resized_height, resized_width = resized_image.shape[:2]

    cv2_imshow(image)
    cv2_imshow(resized_image)

    print(f"Original Image Size: {original_width} x {original_height}")
    print(f"Resized Image Size: {resized_width} x {resized_height}")

    resized_file_name = 'resized_' + file_name
    cv2.imwrite(resized_file_name, resized_image)
else:
    print("No image uploaded.")
```

Crop an image

```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

uploaded = files.upload()

if len(uploaded) > 0:

    file_name = next(iter(uploaded))

    image = cv2.imread(file_name)

    original_height, original_width = image.shape[:2]

    x1 = int(input("Enter the starting X coordinate: "))
    y1 = int(input("Enter the starting Y coordinate: "))
    x2 = int(input("Enter the ending X coordinate: "))
    y2 = int(input("Enter the ending Y coordinate: "))

    cropped_image = image[y1:y2, x1:x2]

    cropped_height, cropped_width = cropped_image.shape[:2]

    cv2_imshow(image)
    cv2_imshow(cropped_image)

    print(f"Original Image Size: {original_width} x {original_height}")
    print(f"Cropped Image Size: {cropped_width} x {cropped_height}")

    cropped_file_name = 'cropped_' + file_name
    cv2.imwrite(cropped_file_name, cropped_image)
else:
    print("No image uploaded.")
```

Read an image dataset and write in suitable directory

```python
import cv2
import os
import shutil
from google.colab import files

print("Upload the source image dataset directory (ZIP file or folder):")
uploaded_source = files.upload()

if len(uploaded_source) > 0:
    source_directory = next(iter(uploaded_source))

    target_directory = "/content/drive/MyDrive/Output"

    if not os.path.exists(target_directory):
        os.makedirs(target_directory)

    if source_directory.endswith(".zip"):
        import zipfile

        with zipfile.ZipFile(source_directory, 'r') as zip_ref:
            zip_ref.extractall(target_directory)

        source_directory = os.path.join(target_directory, os.listdir(target_directory)[0])

    for filename in os.listdir(source_directory):
        if filename.endswith(".jpg") or filename.endswith(".png"):

            source_path = os.path.join(source_directory, filename)
            target_path = os.path.join(target_directory, filename)


            image = cv2.imread(source_path)

            if image is not None:
                cv2.imwrite(target_path, image)
                print(f"Copied: {filename}")
            else:
                print(f"Failed to read image: {filename}")

    print("Image copying complete.")
else:
    print("No source directory uploaded.")
```
