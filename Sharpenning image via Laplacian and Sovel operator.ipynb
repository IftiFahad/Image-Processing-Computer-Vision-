```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

input_folder_path = '/content/drive/MyDrive/dip image sharpenning'
output_folder_path = '/content/drive/MyDrive/dip output'

# Create the output folder if it doesn't exist
os.makedirs(output_folder_path, exist_ok=True)

# List all files in the input folder
image_files = os.listdir(input_folder_path)

for image_file in image_files:
    # Full path to the input image
    input_image_path = os.path.join(input_folder_path, image_file)

    # Read the original image using OpenCV
    original_image = cv2.imread(input_image_path)
    original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
    original_image_gray = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

    # Apply Laplacian filter
    laplacian = cv2.Laplacian(original_image_gray, cv2.CV_64F)
    laplacian = np.uint8(np.absolute(laplacian))

    # Apply Sobel filter in both x and y directions
    sobel_x = cv2.Sobel(original_image_gray, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(original_image_gray, cv2.CV_64F, 0, 1, ksize=3)
    sobel_combined = cv2.magnitude(sobel_x, sobel_y)
    sobel_combined = np.uint8(sobel_combined)

    # Display the images
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 4, 1)
    plt.imshow(original_image_rgb)
    plt.title('Original Image')

    plt.subplot(1, 4, 2)
    plt.imshow(original_image_gray, cmap='gray')
    plt.title('Grayscale')

    plt.subplot(1, 4, 3)
    plt.imshow(laplacian, cmap='gray')
    plt.title('Laplacian Filtered Image')

    plt.subplot(1, 4, 4)
    plt.imshow(sobel_combined, cmap='gray')
    plt.title('Sobel Filtered Image')

    plt.show()

    # Save the processed images to the output folder
    output_laplacian_path = os.path.join(output_folder_path, f'laplacian_{image_file}')
    output_sobel_path = os.path.join(output_folder_path, f'sobel_{image_file}')

    cv2.imwrite(output_laplacian_path, laplacian)
    cv2.imwrite(output_sobel_path, sobel_combined)

# Display a success message
print("Images processed and saved successfully!")
```
