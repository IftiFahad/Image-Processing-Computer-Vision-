```python
from google.colab import drive
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Mount Google Drive
drive.mount('/content/drive')

# Function to apply contrast stretching based on percentiles
def contrast_stretching_percentiles(image_path):
    # Read the image
    image = cv2.imread(image_path, 0)  # Read the image in grayscale mode

    # Calculate the 5th and 95th percentiles of the image
    fifth_percentile = np.percentile(image, 5)
    ninetyfifth_percentile = np.percentile(image, 95)

    # Apply contrast stretching based on the percentiles
    stretched = np.clip((image - fifth_percentile) * (255.0 / (ninetyfifth_percentile - fifth_percentile)), 0, 255)

    # Display the original and stretched images
    plt.figure(figsize=(8, 4))

    # Original Image
    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    # Stretched Image
    plt.subplot(1, 2, 2)
    plt.imshow(stretched, cmap='gray')
    plt.title('Stretched Image')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

# Example: Apply contrast stretching using percentiles to a specific image
image_path = '/content/drive/MyDrive/Output/iron-man1.jpg'  # Replace with the path to your car image
contrast_stretching_percentiles(image_path)
```

```python
from google.colab import drive
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Mount Google Drive
drive.mount('/content/drive')

# Function to apply contrast stretching based on percentiles and show histograms
def contrast_stretching_and_histogram(image_path):
    # Read the image
    image = cv2.imread(image_path, 0)  # Read the image in grayscale mode

    # Calculate the 5th and 95th percentiles of the image
    fifth_percentile = np.percentile(image, 5)
    ninetyfifth_percentile = np.percentile(image, 95)

    # Apply contrast stretching based on the percentiles
    stretched = np.clip((image - fifth_percentile) * (255.0 / (ninetyfifth_percentile - fifth_percentile)), 0, 255)

    # Display the histograms
    plt.figure(figsize=(12, 5))

    # Original Image and Histogram
    plt.subplot(2, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(2, 2, 2)
    plt.hist(image.flatten(), 256, [0, 256], color='r')
    plt.title('Original Histogram')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')

    # Stretched Image and Histogram
    plt.subplot(2, 2, 3)
    plt.imshow(stretched, cmap='gray')
    plt.title('Stretched Image')
    plt.axis('off')

    plt.subplot(2, 2, 4)
    plt.hist(stretched.flatten(), 256, [0, 256], color='r')
    plt.title('Stretched Histogram')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

# Example: Apply contrast stretching using percentiles to a specific image and show histograms
image_path = '/content/drive/MyDrive/Outputan/iron-man1.jpg'  # Replace with the path to your car image
contrast_stretching_and_histogram(image_path)
```

```python
from google.colab import drive
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Mount Google Drive
drive.mount('/content/drive')

# Function to apply Power Law transformation and show the images
def power_law_transformation(image_path, gamma_values):
    # Read the image
    image = cv2.imread(image_path, 0)  # Read the image in grayscale mode

    # Calculate the 5th and 95th percentiles of the image
    fifth_percentile = np.percentile(image, 5)
    ninetyfifth_percentile = np.percentile(image, 95)

    # Apply contrast stretching based on the percentiles
    stretched = np.clip((image - fifth_percentile) * (255.0 / (ninetyfifth_percentile - fifth_percentile)), 0, 255)

    # Display the original and stretched images
    plt.figure(figsize=(12, 8))

    # Original and Stretched Image
    plt.subplot(2, 3, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(2, 3, 2)
    plt.imshow(stretched, cmap='gray')
    plt.title('Contrast Stretched Image')
    plt.axis('off')

    # Apply Power Law transformation with different gamma values
    for i, gamma in enumerate(gamma_values):
        transformed = np.power((stretched / 255.0), gamma) * 255
        plt.subplot(2, 3, i + 4)
        plt.imshow(transformed, cmap='gray')
        plt.title(f'Gamma = {gamma}')
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Example: Apply Power Law transformation to a specific image with different gamma values
image_path = '/content/drive/MyDrive/Outputan/iron-man1.jpg'  # Replace with the path to your car image
gammas = [0.2, 0.5, 1.2, 1.8]
power_law_transformation(image_path, gammas)
```
